```python
#to run
import cv2
from PIL import Image
import numpy as np
import  PIL
import os,glob
import pdf2image
import pytesseract

#image = cv2.imread(a)
def image_ok_hist():

  addon_data = []
  path = os.getcwd()
  path = 'OCC_OG_2GF1A9I_1QR8BL9'
  pdf_files = glob.glob(os.path.join(path, "*.pdf"))
  pdf_files 


  for i in pdf_files:

    try:
      images = pdf2image.convert_from_path(i)
      print(len(images))
      
      dir_name = os.path.basename(i).split('.pdf')[0]
      if not os.path.exists('data '+dir_name):
        os.mkdir('data '+dir_name)

      #start_page = int(input(f"Enter the page number for {i}"))

      for page_num in range(70, len(images)):

        raw_data = open('data '+dir_name+'/page-'+str(page_num+1)+'.csv','w')

        #pil_im = images[page_num] 
        gray = cv2.cvtColor(np.asarray(images[page_num]), cv2.COLOR_BGR2GRAY)
        thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

        # Remove vertical lines
        vertical_kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (2,55))
        detect_vertical = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, vertical_kernel, iterations=2)
        cnts = cv2.findContours(detect_vertical, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        cnts = cnts[0] if len(cnts) == 2 else cnts[1]
        for c in cnts:
            cv2.drawContours(gray, [c], -1, (255,255,255), 9)

        data = pytesseract.image_to_string(gray, lang='eng',config='--psm 6')
        data = data.replace(',','')
        
        raw_data.writelines(data)
        #print(i)

    except Exception as e:
        print(e)
        print(i)
        continue

image_ok_hist()
'''open_raw =  open('raw'+str(page_num)+'.csv','r')
print('read',open_raw.readlines())'''

'''for d in open_raw.readlines():
    addon_data.append(d +'; page '+str(page_num)+'; file '+i)
print(addon_data)'''
```
